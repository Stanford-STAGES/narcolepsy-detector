```python
# with sns.color_palette(sns.diverging_palette(240, 10, n=9)):
g = sns.catplot(
    y='Resolution', 
    x='Accuracy', 
    hue='Features', 
    row='Kernel', 
    data=df, 
    dodge=True, 
    palette='deep',
    zorder=1, 
    alpha=0.5, 
    kind='strip', 
    height=3.25, aspect=1.25
)
g.map_dataframe(
    sns.pointplot, 'Accuracy', 'Resolution', 'Features', 
    dodge=0.8 - 0.8 / 2, 
    join=False,
    palette='dark',
)
(g.set_axis_labels('Accuracy','Resolution (s)')
  .set_titles('{row_name} {row_var}')
  .tight_layout())
g.savefig('notebooks/tmp_row.png', dpi=300)
```
