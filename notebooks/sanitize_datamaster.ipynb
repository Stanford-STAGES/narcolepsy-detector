# Sanitation

## Insert column

```python
df.insert(1, column='ID', value=df['FileID'])
```

## IHC sanitation

```python
df.loc[df['Cohort'] == 'IHC', 'ID'] = df.query('Cohort == "IHC"')['ID'].apply(lambda x: str(x[:5]))
```

```python
(df.query('Cohort == "IHC"')['ID'].apply(lambda x: len(x)).values == 5).any()
```

## KHC

```python
df.loc[df['Cohort'] == 'KHC', 'ID'] = df.query('Cohort == "KHC"')['ID'].apply(lambda x: x.split("_")[0].zfill(8))
```

```python
(df.query('Cohort == "KHC"')['ID'].apply(lambda x: len(x)).values == 8).any()
```
