# Data preparation

```python
df = pd.read_csv('data/narco_features/nrem_frag/nrem_fragmentation.csv', index_col=0)
df['0_norm'] = df['0'].mul(df['Threshold'])
df['0_sqrt'] = df['0'].apply(np.sqrt)
df['0_norm_sqrt'] = df['0'].mul(df['Threshold']).apply(np.sqrt)
df['0_sqrt_norm'] = df['0'].apply(np.sqrt).mul(df['Threshold'])
df.head()
```

```python
sns.lineplot(data=df.query('Resolution == 10'), x='Threshold', y='0_norm', hue='Label', ci=95)
```

```python
sns.relplot(
    data=df, kind='line', x='Threshold', y='0', 
    hue='Label', ci=95, col='Resolution', col_wrap=5,
    facet_kws={'sharey': False, 'sharex': True}
)
```

```python
sns.relplot(
    data=df, kind='line', x='Threshold', y='0_norm', 
    hue='Label', ci=95, col='Resolution', col_wrap=5,
    facet_kws={'sharey': False, 'sharex': True}
)
```

```python
sns.relplot(
    data=df, kind='line', x='Threshold', y='0_norm_sqrt', 
    hue='Label', ci=95, col='Resolution', col_wrap=5,
    facet_kws={'sharey': False, 'sharex': True}
)
```

```python
sns.relplot(
    data=df, kind='line', x='Threshold', y='0_sqrt_norm', 
    hue='Label', ci=95, col='Resolution', col_wrap=5,
    facet_kws={'sharey': False, 'sharex': True}
)
```

```python
sns.relplot(
    data=df, kind='line', x='Threshold', y='0_sqrt', 
    hue='Label', ci=95, col='Resolution', col_wrap=5,
    facet_kws={'sharey': False, 'sharex': True}
)
```
