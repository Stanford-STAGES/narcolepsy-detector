# Data preparation

```python
df = pd.read_csv('data/narco_features/sorem/sorem.csv', index_col=0)
df['0_norm'] = df['0'].mul(df['Threshold'])
df['1_norm'] = df['1'].mul(df['Threshold'])
df['1_sqrt'] = df['1'].apply(np.sqrt)
df['1_norm_sqrt'] = df['1'].mul(df['Threshold']).apply(np.sqrt)
df['1_sqrt_norm'] = df['1'].apply(np.sqrt).mul(df['Threshold'])
df.head()
```
